# Chargement de la table "sujet_tele"
Cette table est expose le temps de diffusion par chaîne de différentes thématiques

```python
import requests
url_sujet_tele = "https://static.data.gouv.fr/resources/classement-thematique-des-sujets-de-journaux-televises-janvier-2000-decembre-2020/20241015-124725/ina-barometre-jt-tv-donnees-quotidiennes-2000-2020-nbre-sujets-durees-202410.csv"
requests.get(url_sujet_tele)
# Vérification que l'url fonctionne
```

```python
import pandas as pd

req = requests.get(url_sujet_tele)
print(req.content[:20])
# Observe le format des données
```

```python
# Définission du nom des colonnes
colonnes = ["Date", "Chaîne","Vide", "Thématique", "Nb_sujets", "Duree_sec"]
```

```python
with open('temp.csv', 'w', encoding='latin-1') as f:
    f.write(req.text)
sujet_tele = pd.read_csv('temp.csv', sep=';', encoding='latin-1', header=None, names=colonnes) 
# latin permet d'encoder les accents
```

```python
sujet_tele.head()
```

```python
sujet_tele["Thématique"].unique()
```

```python
# Vérification que la colonne "Vide" ne contient pas de valeurs non nulles
sujet_tele['Vide'].notna().any()
```

```python
# Supprime la colonne "Vide"
sujet_tele = sujet_tele.drop(columns=['Vide'])
sujet_tele.head()
```

# Chargement de la base "parite"

```python
req = requests.get(url_parite)
print(req.content[:20])
```
